- La siguiente celda contiene un ejemplo de cómo realizar un *boxplot*. ¿Qué representa este tipo de gráficas? Realiza una gráfica *boxplot* con la subbase del punto anterior. 


```python
import numpy as np 
import pandas as pd


np.random.seed(12345)
df = pd.DataFrame(np.random.randn(11, 4), columns=['Col1', 'Col2', 'Col3', 'Col4'])
boxplot = df.boxplot(column=['Col1', 'Col2', 'Col3'])

```

Aqui lo que podemos ver es que se esta generando numero random con la distribucion estandard normal. Con una "seed" de numeros random lo cual hace que los numero random que se generen sean siemrpe los mismos y sea la misam grafica siempre. El codigo np.random.randn nos va a dar numeros random en una distribucion estandard normal donde hay 4 columans y 10 renglones por columnas. El boxplot nos enseña basicamente el rango que va de cada columna en los numero, mostrando el máximo y el minimo, la línea verde nos enseña la mediana y los circulos fuera son los "outliers".

```python
subbase20.boxplot(["salary_in_usd"])
```

Teniendo algunos outliers como de 325,000 dolares, 380,000 y de 405,000. Esto nos dice que lo normal de los mejores 20 puestos mejor pagados en el 2022 si es que queremos apuntar a ellos en la parte laboral son alrededor de 230,000 dolares anuales, algo bastante bien para vivir.

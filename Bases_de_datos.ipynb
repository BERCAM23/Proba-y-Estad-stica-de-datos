```python
import numpy as np 
import pandas as pd


np.random.seed(12345)
df = pd.DataFrame(np.random.randn(11, 4), columns=['Col1', 'Col2', 'Col3', 'Col4'])
boxplot = df.boxplot(column=['Col1', 'Col2', 'Col3'])

```

Aqui lo que podemos ver es que se esta generando numero random con la distribucion estandard normal. Con una "seed" de numeros random lo cual hace que los numero random que se generen sean siemrpe los mismos y sea la misam grafica siempre. El codigo np.random.randn nos va a dar numeros random en una distribucion estandard normal donde hay 4 columans y 10 renglones por columnas.
